# Minimum Spanning Tree Problem

## Import Libraries

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import random
```

## Generate Radom Graph

```python
n = 7
prob = 0.5

G = nx.erdos_renyi_graph(n, prob, seed=None, directed=False)

for (u, v) in G.edges():
    G.add_edge(u,v,weight=random.randint(0, 9))
  
layout = nx.kamada_kawai_layout(G)

plt.rcParams["figure.figsize"] = (10, 10)

nx.draw(G, pos=layout, with_labels=True) # default layout
```

## Solving using PRIMS and KRUSKAL algorithms

```python
KR = nx.minimum_spanning_tree(G, algorithm="kruskal")
print("Kruskal's Algorithm \n",sorted(KR.edges(data=True)))

PR = nx.minimum_spanning_tree(G, algorithm="prim")
print("Prim's Algorithm \n",sorted(PR.edges(data=True)))
```

## Draw a Kruskal & Prim's Graphs

```python
plt.rcParams["figure.figsize"] = (20, 10)

fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Graphic Answer')

plt.sca(ax1)
plt.title("Prim")
nx.draw(PR, pos=layout, with_labels=True) # default layout

plt.sca(ax2)
plt.title("Kruskal")
nx.draw(KR, pos=layout, with_labels=True) # default layout
```
